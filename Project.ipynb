```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols =['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']
df =df[cols]
```

```python
df
```

```python
df.rename(columns={'ActivityDate':'date'},inplace=True)
```

```python
df
```

```python
df['TotalMinutes'] = df.VeryActiveMinutes + df.FairlyActiveMinutes + df.LightlyActiveMinutes + df.SedentaryMinutes
```

```python
df
```

```python
df['TotalHours'] = round(df.TotalMinutes/60)
```

```python
df.date= pd.to_datetime(df.date)
```

```python
import datetime as dt
df['DayofWeek'] = df.date.dt.day_name()
```

```python
plt.figure(figsize =((6,4)))
plt.hist(df.DayofWeek,bins=7, color = 'lightskyblue',width = 0.6)
plt.grid(True)

plt.xlabel('day of week')
plt.ylabel('freq')

plt.title('freq by day of week')
plt.show()
```

```python
plt.figure(figsize =((8,6)))
plt.scatter(df.TotalSteps, df.Calories, c=df.Calories)

median_steps =7405
median_calories=2134

plt.axhline(median_calories, color='blue',label ='mediam of calories')
plt.axvline(median_steps, color='red',label ='mediam of steps')

plt.xlabel('Steps')
plt.ylabel('Calories')

plt.title('Calores by total steps')
plt.legend()
plt.show()
```

```python
plt.figure(figsize =((8,6)))
plt.scatter(df.TotalHours, df.Calories, c=df.Calories)

median_hours =24
median_calories=2134

plt.axhline(median_calories, color='blue',label ='mediam of calories')
plt.axvline(median_hours, color='red',label ='mediam of hours')

plt.xlabel('hours')
plt.ylabel('Calories')

plt.title('Calores by total steps')
plt.legend()
plt.show()
```
